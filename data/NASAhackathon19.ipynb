```python
import pandas as pd
AOD_raw  = pd.read_csv('/content/gdrive/My Drive/NASA/taiwan.csv') #, encoding="utf-8"
#PM_raw = []
PM18_raw = pd.read_csv('/content/gdrive/My Drive/NASA/p1_utf16.csv', encoding="utf-16")
#PM_raw = pd.read_csv('/content/gdrive/My Drive/NASA/桃園五年空汙資料/107年中壢站_20190315.csv', encoding='mbcs') #ansi encoding
# encoding="UTF-8-BOM"

```

```python
import math

PM = PM18_raw
l_PM = PM.shape[0]
#print(PM.columns)
#print(PM)

PM25_18 = {}
for i in range(l_PM) :
  if PM["測項"][i] == "PM2.5":
    sum24 = 0.0
    n_sum = 0.0
    for i24 in range(24):
      idx = str(i24)
      if i24<10 :
        idx = '0'+idx
      #print(PM[idx][i])
      print(str(PM[idx]))
      if str(PM[idx][i]).isdigit():
      #if not math.isnan(PM[idx][i]):
        n_sum += 1
        sum24 += float(PM[idx][i])
    PM25_18[PM["日期"][i]] = sum24/n_sum
```

```python
import csv
csv_file = "/content/gdrive/My Drive/NASA/PM25_18.csv"
csv_columns = ['date', 'avg PM2.5']
csv_data = [{'date': key , 'avg PM2.5':PM25_18[key] } for key in PM25_18]

print(csv_data)

try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in csv_data:
            writer.writerow(data)
except IOError:
    print("I/O error") 
```
