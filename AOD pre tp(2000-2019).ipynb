```python
import pandas as pd
AOD_raw  = pd.read_csv('./data_tp/Taipei_CWB.csv') #, encoding="utf-8"
AOD_raw.shape
```

```python
AOD_raw.columns
```

```python
AODr = [AOD_raw['AERONET Version 3; '][i].split(',') for i in range(AOD_raw.shape[0])]
#AOD[5]
```

```python
AOD = AODr[6:] #0=location 1=date 2=time 3=dayofyear 37(4:40), all82, 35, 6
#AOD18[0][4:40]
AODd = {}
for i in range(len(AOD)):
    day = AOD[i][0][0:2]
  #print(day)
    month = AOD[i][0][3:5]
    year = AOD[i][0][6:10]
    date = str(year) + '/' + str(month) + '/' + str(day)
    AODd[date] = [float(AOD[i][k]) for k in range(3, 41)] #take 3-40 as feature
```

```python
n_feature = 41-3
avgs = []
for i in range(n_feature):
    summation = 0.0
    cnt = 0.0
    for key in AODd:
        if not AODd[key][i] == -999.:
            cnt += 1
            summation +=  AODd[key][i]
    avg = summation/cnt if cnt != 0 else 0
    avgs.append(avg)
    for key in AODd:
        if AODd[key][i] == -999.:
            AODd[key][i] = avg
  
```

```python
import csv
csv_file = "./data_tp/AOD.csv"
csv_columns = ['date', 'data']

try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(csv_columns)
        for key, val in AODd.items():
            writer.writerow([key,val])
except IOError:
    print("I/O error") 
```
